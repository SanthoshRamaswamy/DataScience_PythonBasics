```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
beml = pd.read_csv("/content/BEML.csv")
```

```python
beml
```

```python
beml = beml[['Date','Close']]
```

```python
beml = beml.set_index(pd.DatetimeIndex(beml['Date']))
```

```python
beml
```

```python
beml['gain'] = beml.Close.pct_change(periods=1)
```

```python
beml
```

```python
beml.removena()
```

```python
beml.dropna()
```

```python
#calculating interval
beml_CI = stats.norm.interval(0.95, loc=beml.gain.mean(), scale=beml.gain.std())
```

```python
print("Gain at 95% CI is ", np.round(beml_CI, 4))
```

```python
glaxo = pd.read_csv("/content/GLAXO.csv")
```

```python
glaxo
```

```python
#taking only date and close price columns
glaxo = glaxo[['Date', 'Close']]
```

```python
glaxo
```

```python
#changing the index to Date using pd.DatetimeIndex() function
glaxo = glaxo.set_index(pd.DatetimeIndex(glaxo['Date']))
```

```python
glaxo
```

```python
# calculating gain by using pct_change() method
glaxo['gain'] = glaxo.Close.pct_change(periods=1)
```

```python
glaxo
```

```python
#droping NA 
glaxo.dropna()
```

```python
#finding gain of GLAXO at 95% CI
glaxo_CI = stats.norm.interval(0.95, loc=glaxo.gain.mean(), scale=glaxo.gain.std())
print('Gain of GLAXO at 95% CI is ', np.round(glaxo_CI, 4))
```
